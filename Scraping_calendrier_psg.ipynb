```python
import re 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.planetepsg.com/calendrier-2023-2024.html"
requesting = requests.get(url)
soup  = BeautifulSoup(requesting.content, "html.parser")
liste=soup.select("html body div.col-xs-12 div.container div.row div.col-xs-12.col-sm-12.col-md-8 div.row.Matches_Listing")
```

```python
def parser(x):
    return x.get_text().strip().split("\n")

df_ = pd.DataFrame(list(map(lambda x:parser(x), liste)))
df_
```

```python
#On recupère les matchs non joués 
df = df_.iloc[26:]
df
```
